```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.random(100)
print(x)
```

```python
plt.hist(x, 10)

plt.xlabel("x value")
plt.ylabel("Numbers of elements")
plt.show()
```

```python
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 10000)
plt.hist(s,10)

plt.show()
```

```python
ss = plt.hist(s, 10)
print(ss)
```

```python
bins = np.linspace(-0.5, 0.5, 21, endpoint=True)
print(bins)
plt.hist(s,bins)

plt.show()
```
